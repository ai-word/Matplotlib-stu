MATLAB风格接口:有状态的,简单，和当前图形绑定，但是回到之前的难度大

面向对象接口:显式的绑定所有图形，方便切换

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,100)
```

```python
# ax是一个包含两个Axes对象的数组
fig, ax = plt.subplots(2)
# 在每个对象上调用plot()方法 
ax[0].plot(x, np.sin(x)) 
ax[1].plot(x, np.cos(x))
ax[0].plot(x,np.cos(x));
```
